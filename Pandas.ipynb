```python
import pandas as pd
import matplotlib.pyplot as plt

air_quality = pd.read_csv("air_quality_no2.csv",index_col=0,parse_dates=True)
```

```python
air_quality.head()
```

```python
# create plots in pandas

air_quality.plot()
```

```python
#plotting the column of the particular datatable.
#In plot we have scatter,area,box.

air_quality["station_paris"].plot()
```

```python
# No2 values measured in london and paris.

air_quality.plot.scatter(x = "station_london",y = "station_paris",alpha = 1)
```

```python
[method_name for method_name in dir(air_quality.plot) if not method_name.startswith("_")]
```

```python
#box

air_quality["station_london"].plot()
```

```python
#box

air_quality.plot.box()
```

```python
#area for sub_plots with area function consists parameters figsize and subplots

axs = air_quality.plot.area(figsize=(12,4), subplots = True)
```

```python
# Create an empty matplotlib Figure and Axis where figsize x = 12cms and y = 4 cms.
fig, axs = plt.subplots(figsize=(12,4))
```

```python
#use pandas to put the area plot on the prepared Figures/Axes

air_quality.plot.area(ax=axs)
```

```python
#do any matplotlib customization you like

axs.set_ylabel("NO$_2$ concentration")
```

```python
# Save the Figure/Axes using the existing matplotlib method

fig.savefig("no2_concentrations.png")
```
